```python
import hdbscan
import numpy as np

# Example data: Random points in 2D space
data = np.random.rand(100, 2)

# Perform HDBSCAN clustering, enabling minimum spanning tree generation
clusterer = hdbscan.HDBSCAN(min_cluster_size=10, gen_min_span_tree=True) # Set gen_min_span_tree to True
cluster_labels = clusterer.fit_predict(data)

print(cluster_labels)  # Cluster labels or -1 for noise

clusterer.minimum_spanning_tree_.plot(edge_cmap='viridis',
                                      edge_alpha=0.6,
                                      node_size=80,
                                      edge_linewidth=2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import hdbscan

# Tạo dữ liệu giả
data = np.array([
    [0, 0], [0.1, 0.1], [0.2, 0.2],
    [5, 5], [5.1, 5.1], [5.2, 5.2],
    [10, 10]
])

# Chạy HDBSCAN
clusterer = hdbscan.HDBSCAN(min_cluster_size=2)
clusterer.fit(data)
```

```python
# Vẽ Single Linkage Tree
clusterer.single_linkage_tree_.plot()
plt.show()
```

```python
condensed_tree = clusterer.condensed_tree_.to_pandas()
print(condensed_tree.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import hdbscan

# Tạo dữ liệu mẫu
data = np.array([
    [0, 0], [1, 1], [2, 2],
    [4.5, 4.5], [6, 6], [6.5, 6.5],
    [7 , 7]
])

# Chạy HDBSCAN
clusterer = hdbscan.HDBSCAN(min_cluster_size=2)
clusterer.fit(data)

# Vẽ Single Linkage Tree
clusterer.single_linkage_tree_.plot()

# THÊM: Sau khi plot xong mới lấy trục hiện tại để annotate
ax = plt.gca()

# Annotate các giá trị Lambda
for child, parent, lambda_val, child_size in clusterer.single_linkage_tree_.to_numpy():
    x = (child + parent) / 2  # x là trung điểm giữa 2 node
    y = lambda_val            # y là giá trị lambda
    ax.text(x, y + 0.5, f"λ={lambda_val:.2f}", ha='center', va='bottom', fontsize=8, color='red')

plt.title('Single Linkage Tree with Lambda values')
plt.xlabel('Points/Clusters')
plt.ylabel('Lambda (1/Distance)')
plt.show()
```

```python
condensed_tree = clusterer.condensed_tree_.to_pandas()
condensed_tree['distance'] = 1 / condensed_tree['lambda_val']
print(condensed_tree)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import hdbscan
from scipy.cluster.hierarchy import dendrogram, linkage

# Bước 1: Tạo một dataset ngẫu nhiên
X, _ = make_blobs(n_samples=15, centers=4, random_state=20)

# Bước 2: Áp dụng HDBSCAN
clusterer = hdbscan.HDBSCAN(min_cluster_size=2)
clusterer.fit(X)

# Bước 3: Tạo dendrogram
# Sử dụng linkage để tính toán các liên kết hierarchial giữa các cụm
linked = linkage(X, 'ward')

# Vẽ dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title('Dendrogram for HDBSCAN clustering')
plt.xlabel('Index')
plt.ylabel('Distance')
plt.show()

# # Hiển thị kết quả phân cụm từ HDBSCAN
# plt.scatter(X[:, 0], X[:, 1], c=clusterer.labels_, cmap='viridis')
# plt.title('HDBSCAN Clustering Results')
# plt.xlabel('Feature 1')
# plt.ylabel('Feature 2')
# plt.show()
```
